```python
import pandas as pd

df = pd.read_csv('author_count.csv', encoding='utf-16', sep=',')
df['frequency'] = df['count'] / df['count'].sum()
df.describe().apply(lambda s: s.apply(lambda x: format(x, '.0f') if x > 1 else format(x, '.6f')))
```

```python
df.sort_values(by='count', ascending=False, inplace=True)
df.head(10)
```

```python
ignore_authors = {'Various', 'Anonymous'}
```

```python
df_nostopwords = df[~df['author'].isin(ignore_authors)].copy()
df_nostopwords['frequency'] = df_nostopwords['count'] / df_nostopwords['count'].sum()

df_nostopwords.sort_values(by='count', ascending=False, inplace=True)
df_nostopwords.head(10)
```
